## Simple Convolutional VAE

Here $\mathbf{x} = (x_k)_{k \in [1, 784]}$

In this case, we use as a loss function :
$$L_{TOTAL} = L_{RECONSTRUCTION} + L_{REGULARISATION}$$

Where $L_{RECONSTRUCTION} = - \sum_k{x_k log(\hat{x}_k) + (1-x_k)log(1 - \hat{x}_k) }$ is the *binary cross-entropy*. We usually use the binary crossentropy when outputs are either 0 ior 1 but we can still use it as a loss function when the inputs $x_k$ (i.e. the pixels) are valued valued between zero and one. The loss function is still minimised for $\forall k, \hat{x}_k = x_k$ but it will never be 0.
, and are close to zero or one (white digits written on black background.

For real-valued inputs, we can use $L_{RECONSTRUCTION} = MSE$ or $MAE$ with linear activation function at the output layer.

```python
# NE PAS UTILISER keras MAIS tf.keras

from CVAE import Sampling, VAE
```

```python
# define encoder and decoder with the Functional API
# define the VAE with the Model Subclassing method

original_dim = 784
latent_dim = 2
latent_side = int(np.sqrt(latent_dim))
kernel_size = 3
dropout = 0
batchnorm = False
n_filters = 4

encoder_inputs = keras.Input(shape=(28, 28, 1))
x = layers.Conv2D(32, 3, input_shape=(28, 28, 1), activation="relu", strides=2, padding="same")(encoder_inputs)
x = layers.Conv2D(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Flatten()(x)
x = layers.Dense(32, activation="relu")(x)
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
z = Sampling()([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")

# encoder.summary()

latent_inputs = keras.Input(shape=(latent_dim,))
x = layers.Dense(7 * 7 * 64, activation="relu")(latent_inputs)
x = layers.Reshape((7, 7, 64))(x)
x = layers.Conv2DTranspose(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2DTranspose(32, 3, activation="relu", strides=2, padding="same")(x)
decoder_outputs = layers.Conv2DTranspose(1, 3, activation="sigmoid", padding="same")(x)
decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
```

```python
(x_train, _), (x_test, _) = keras.datasets.mnist.load_data()

x_train = np.expand_dims(x_train, -1).astype("float32") / 255 # add width axis
x_test = np.expand_dims(x_train, -1).astype("float32") / 255

vae = VAE(encoder, decoder, (28, 28, 1))
vae.compile(optimizer=keras.optimizers.Adam())
vae.fit(x_train, epochs=5, batch_size=128)
```

### Topology of the latent space

```python
# latent space, from keras.io
def plot_label_clusters(encoder, decoder, data, labels):
    # display a 2D plot of the digit classes in the latent space
    # print(data.shape)
    z_mean, _, _ = encoder.predict(data)

    plt.figure(figsize=(6, 6))
    plt.scatter(z_mean[:, 0], z_mean[:, 1], c=labels)
    plt.colorbar()
    plt.xlabel("z[0]")
    plt.ylabel("z[1]")
    plt.show()

x_test = x_test.reshape((x_test.shape[0], 28, 28, 1))

print(x_test.shape)

plot_label_clusters(encoder, decoder, x_test, y_test)
```

```python
# latent space, from keras.io
def plot_latent(encoder, decoder):
    # display a n*n 2D manifold of digits
    n = 30
    digit_size = 28
    scale = 2
    figsize = 10
    figure = np.zeros((digit_size * n, digit_size * n))
    # linearly spaced coordinates corresponding to the 2D plot
    # of digit classes in the latent space
    grid_x = np.linspace(-scale, scale, n)
    grid_y = np.linspace(-scale, scale, n)[::-1]

    for i, yi in enumerate(grid_y):
        for j, xi in enumerate(grid_x):
            z_sample = np.array([[xi, yi]])
            x_decoded = decoder.predict(z_sample)
            digit = x_decoded[0].reshape(digit_size, digit_size)
            figure[
                i * digit_size : (i + 1) * digit_size,
                j * digit_size : (j + 1) * digit_size,
            ] = digit

    plt.figure(figsize=(figsize, figsize))
    start_range = digit_size // 2
    end_range = n * digit_size + start_range + 1
    pixel_range = np.arange(start_range, end_range, digit_size)
    sample_range_x = np.round(grid_x, 1)
    sample_range_y = np.round(grid_y, 1)
    plt.xticks(pixel_range, sample_range_x)
    plt.yticks(pixel_range, sample_range_y)
    plt.xlabel("z[0]")
    plt.ylabel("z[1]")
    plt.imshow(figure, cmap="Greys_r")
    plt.show()

plot_latent(encoder, decoder)
```

```python
# if x_test had dim=(?, 28, 28)
reconst_loss_norm = np.array([np.sum((xi_test - predictions_i) ** 2) for xi_test, predictions_i in zip(x_test, predictions)])
reconst_loss_anom = np.array([np.sum((xi_test - predictions_i) ** 2) for xi_test, predictions_i in zip(anomalies, anomalies_pred)])

print(reconst_loss_norm[0])

fig, ax = plt.subplots(1, 1)
ax.hist(reconst_loss_norm, bins=20, label="Original images")
ax.hist(reconst_loss_anom, bins=20,label="Images with anomalies")
ax.legend()
fig.suptitle("Reconstruction loss on original and anormal images")

plt.show()
```
